```python
import pandas as pd

# Veriyi içeri alıyoruz (Dosyada veriler noktalı virgülle ayrıldığı için sep=';' ekledik)
df = pd.read_csv('zara clothes.csv', sep=';')

# Tablonun ilk 5 satırını ekrana dökerek verimizi inceliyoruz
df.head()
```

```python
# Ürünlerin konumuna (Reyon başı vs. Koridor) göre ortalama satış rakamlarını buluyoruz
df.groupby('Product Position')['Sales Volume'].mean().round()
```

```python
# Hem ürün konumuna hem de indirim durumuna göre ortalama satışları grupluyoruz
df.groupby(['Product Position', 'Promotion'])['Sales Volume'].mean().round()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Veriyi içeri alıyoruz (Hafıza sıfırlanmasına karşı önlem)
df = pd.read_csv('zara clothes.csv', sep=';')

# 2. Grafiği çizdiriyoruz
plt.figure(figsize=(8, 5))
sns.set_theme(style="white")

# Digital & Fresh paletimiz (Soft Lavanta ve Su Yeşili)
colors = ["#B5A8D5", "#A8DADC"]

ax = sns.barplot(
    x="Product Position",
    y="Sales Volume",
    hue="Promotion",
    data=df,
    palette=colors,
    edgecolor="white",
    linewidth=2
)

# Yazı tipleri ve renklerini soft koyu griye çekiyoruz
plt.title("Vitrinleri Aslında Kim Yönetiyor? - Zara Satış Analizi", fontsize=13, pad=15, color='#555555', fontweight='bold')
plt.xlabel("Ürün Konumu", fontsize=10, color='#777777')
plt.ylabel("Ortalama Satış", fontsize=10, color='#777777')

# Lejant ve Çerçeve temizliği
plt.legend(title="İndirim", frameon=False, loc="upper right")
sns.despine(left=True, bottom=True)

plt.tight_layout()
plt.show()
```
