```python
# 1. Veriyi baştan, tertemiz bir şekilde okutuyoruz (Hafızayı sıfırladık)
dosya_yolu = '/content/Women_s_E-Commerce_Clothing_Reviews_1594_1.csv'
df = pd.read_csv(dosya_yolu, sep=';')

# 2. Gereksiz sütunu atıyoruz
df = df.drop(columns=['Unnamed: 0'], errors='ignore')

# 3. ÖNCE sütun isimlerindeki noktaları alt çizgi yapıyoruz
df.columns = df.columns.str.replace('.', '_', regex=False)

# 4. ŞİMDİ yeni ismiyle (Review_Text) içinde metin olmayan boş satırları siliyoruz
df = df.dropna(subset=['Review_Text'])

# Sonucu görelim
print(f"Temizlik sonrası kalan satır sayısı: {len(df)}")
```

```python
# Gerekli görselleştirme kütüphanelerini çağırıyoruz
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

# Bütün yorumları tek bir devasa metin (string) haline getiriyoruz
butun_yorumlar = " ".join(yorum for yorum in df['Review_Text'].astype(str))

# Kelime bulutunun tasarım ayarlarını yapıyoruz
wordcloud = WordCloud(width=800, height=400,
                      background_color='white',
                      stopwords=STOPWORDS,  # the, and, is gibi bağlaçları eler
                      colormap='magma').generate(butun_yorumlar)

# Grafiği ekrana çizdiriyoruz
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off') # Çirkin x ve y ekseni çizgilerini gizliyoruz
plt.title("Kadın Giyim Yorumlarında En Çok Geçen Kelimeler", fontsize=16)
plt.show()
```

```python
# 1. Sadece 1 ve 2 yıldız alan (kötü) yorumları filtreliyoruz
kotu_yorumlar_df = df[df['Rating'] <= 2]

# 2. Bu kötü yorumların hepsini tek bir metin haline getiriyoruz
kotu_metin = " ".join(yorum for yorum in kotu_yorumlar_df['Review_Text'].astype(str))

# 3. Kötü yorumlar için kelime bulutunu oluşturuyoruz
# 'Reds' colormap'i ile olumsuz havayı görsele yansıtıyoruz
wordcloud_kotu = WordCloud(width=800, height=400,
                           background_color='white',
                           stopwords=STOPWORDS,
                           colormap='Reds').generate(kotu_metin)

# 4. Ekrana çizdirme kısmı
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud_kotu, interpolation='bilinear')
plt.axis('off')
plt.title("Düşük Puan Veren (1-2 Yıldız) Müşterilerin En Çok Kullandığı Kelimeler", fontsize=16)
plt.show()

# Kaç tane kötü yorum olduğunu da görelim
print(f"Analiz edilen kötü yorum sayısı: {len(kotu_yorumlar_df)}")
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# 1. Yaşları daha anlamlı gruplara (bin) ayırıyoruz
yas_sinirlari = [0, 25, 35, 45, 55, 65, 100]
grup_isimleri = ['18-25', '26-35', '36-45', '46-55', '56-65', '65+']
df['Age_Group'] = pd.cut(df['Age'], bins=yas_sinirlari, labels=grup_isimleri)

# 2. Hangi yaş grubundan kaç yorum geldiğini sayıyoruz
yas_dagilimi = df['Age_Group'].value_counts().sort_index()

# 3. Bar grafiğini çizdiriyoruz
plt.figure(figsize=(10, 6))
yas_dagilimi.plot(kind='bar', color='#4C72B0', edgecolor='black')

# Grafiğin makyaj ayarları (Başlık, eksen isimleri vb.)
plt.title("Moda E-Ticaretinde Yorum Yapanların Yaş Dağılımı", fontsize=16, fontweight='bold')
plt.xlabel("Yaş Grupları", fontsize=12)
plt.ylabel("Yorum Sayısı", fontsize=12)
plt.xticks(rotation=0, fontsize=11)
plt.grid(axis='y', linestyle='--', alpha=0.7) # Arkaya şık bir ızgara ekliyoruz
plt.show()
```
